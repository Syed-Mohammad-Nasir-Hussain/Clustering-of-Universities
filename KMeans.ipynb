```python
a1=a.iloc[:,1:] # Removing University names column
```

```python
from sklearn.cluster import KMeans # Required Library
```

```python
# Storing distortion
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(a1)
    distortions.append(kmeanModel.inertia_)
```

```python
# Plotting elbow curve
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
